```python
import pandas as pd

url = 'https://raw.githubusercontent.com/BDonadelli/Codigos-em-financas/main/data/Cart_Ibov.csv'
df = pd.read_csv(url)
df
```

```python
import yfinance as yf
tickers = (df['Código']+'.SA').to_list()
preco = yf.download(tickers,period='3y',auto_adjust=True)['Close']
preco
```

```python
preco.columns[preco.isnull().any()]
```

```python
preco.drop(columns=['ALOS3.SA', 'IGTI11.SA', 'RAIZ4.SA'],inplace=True)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm
```

```python
scaler = StandardScaler() # normalização padrão 
ret = preco.pct_change().dropna()
rets = scaler.fit_transform(ret)
retornos = pd.DataFrame(rets , index=ret.index, columns=ret.columns)
retornos
```

```python
n = 5
pca = PCA(n_components = n)             # Inicializando o PCA com n componentes principais
pca.fit(retornos)                   
pct = pca.explained_variance_ratio_     # Obtendo a razão da variância explicada por cada componente principal
pca_components = pca.components_        # Obtendo os componentes principais
```

explained_variance_ratio_ é um atributo do objeto PCA que contém a proporção da variância total explicada por cada um dos componentes principais selecionados. Isso é útil para entender quanta informação (variância) dos dados originais está sendo capturada por cada componente principal. O resultado é uma matriz de n_components elementos onde cada elemento representa a variância explicada por um componente principal.

components_ é um atributo do objeto PCA que contém os vetores dos componentes principais. Esses vetores representam a direção no espaço original dos dados que maximiza a variância. O resultado é uma matriz onde cada linha corresponde a um dos componentes principais (n_components linhas, no caso), e cada coluna corresponde a uma das características originais dos dados.


```python
import matplotlib.pyplot as plt
import numpy as np

# Calculando a variância acumulada
var_acumulada = np.cumsum(pct)

# Configuração do gráfico
plt.figure(figsize=(14, 7))

# Gráfico de Barras para a Variância Explicada
plt.subplot(1, 2, 1)
plt.bar(range(1, len(pct) + 1), pct, alpha=0.7, align='center', color='blue')
plt.xlabel('Componentes Principais')
plt.ylabel('Variância Explicada (%)')
plt.title('Contribuição de Cada Componente Principal')
plt.xticks(range(1, len(pct) + 1))

# Gráfico de Linhas para a Variância Acumulada
plt.subplot(1, 2, 2)
plt.plot(range(1, len(var_acumulada) + 1), var_acumulada, marker='o', linestyle='--', color='red')
plt.xlabel('Componentes Principais')
plt.ylabel('Variância Acumulada (%)')
plt.title('Variância Acumulada por Componentes Principais')
plt.xticks(range(1, len(var_acumulada) + 1))
plt.ylim(0, 1.05)

# Exibindo os gráficos
plt.tight_layout()
plt.show()
```

```python
# Transformando os dados para obter os scores dos componentes principais
pca_scores = pca.transform(retornos)

# Convertendo os scores para um DataFrame para facilitar a manipulação
pca_scores_df = pd.DataFrame(pca_scores, columns=[f'C{i+1}' for i in range(n)])

# Plotando os scores dos componentes principais ao longo do tempo
plt.figure(figsize=(14, 7))
for i in range(n):
    plt.plot(pca_scores_df.index, pca_scores_df[f'C{i+1}'], label=f'Componente {i+1}')
plt.xlabel('Data')
plt.ylabel('Score do Componente Principal')
plt.title('Scores dos Componentes Principais ao Longo do Tempo')
plt.legend()
```

```python
X = np.asarray(retornos)
components_returns = X.dot(pca_components.T)

components_returns = pd.DataFrame(
    columns=[f'C{i+1}' for i in range(n)] ,
    index=retornos.index,
    data=components_returns
)

components_returns
```
