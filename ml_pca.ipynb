```python
import pandas as pd
url = 'https://raw.githubusercontent.com/BDonadelli/Codigos-em-financas/main/data/Cart_Ibov.csv'
df = pd.read_csv(url)
df
```

```python
import yfinance as yf
tickers = (df['Código']+'.SA').to_list()
preco = yf.download(tickers,period='1y',auto_adjust=True)['Close']
preco
```

```python
preco.columns[preco.isnull().any()].to_list()
```

```python
dropes = preco.columns[preco.isnull().any()].to_list()
```

```python
preco.drop(columns=dropes,inplace=True)
```

```python
import statsmodels.api as sm
```

```python
# padronizar a variação percentual de uma série 

from sklearn.preprocessing import StandardScaler #uma ferramenta de pré-processamento usada para padronizar recursos, ou seja, reescalá-los para que tenham média zero e desvio padrão 1.
scaler = StandardScaler()  #instância do StandardScaler, que será usada para transformar os dados.
ret = preco.pct_change().dropna()
rets = scaler.fit_transform(ret) # ajusta o StandardScaler nos dados e aplica a transformação, reescalando os dados de modo que cada coluna tenha média zero e desvio padrão igual a 1. A saída é um array numpy com os dados transformados.
retornos = pd.DataFrame(rets , index=ret.index, columns=ret.columns)
retornos
```

```python
'''
reduzir a dimensionalidade dos dados em retornos, ou seja, os valores padronizados obtidos previamente. 
O PCA é útil para identificar as principais direções (componentes) em que os dados variam, permitindo 
resumir os dados de forma mais compacta. 
'''
from sklearn.decomposition import PCA

n = 5 # o PCA tentará reduzir os dados para um espaço de 5 dimensões , capturando o máximo possível de variabilidade com essas 5 componentes.
pca = PCA(n_components = n)             
pca.fit(retornos)                   
pct = pca.explained_variance_ratio_      # a proporção de variância explicada por cada um dos componentes principais. 
pca_components = pca.components_         # contém as direções dos componentes principais.
```

explained_variance_ratio_ é um atributo do objeto PCA que contém a proporção da variância total explicada por cada um dos componentes principais selecionados. Isso é útil para entender quanta informação (variância) dos dados originais está sendo capturada por cada componente principal. O resultado é uma matriz de n_components elementos onde cada elemento representa a variância explicada por um componente principal.

components_ é um atributo do objeto PCA que contém os vetores dos componentes principais. Esses vetores representam a direção no espaço original dos dados que maximiza a variância. O resultado é uma matriz onde cada linha corresponde a um dos componentes principais (n_components linhas, no caso), e cada coluna corresponde a uma das características originais dos dados.


```python
''' visualiza a variância explicada e acumulada pelas componentes principais após a aplicação do PCA, 
ajudando a entender como cada componente contribui para a explicação dos dados e quantos componentes 
são necessários para capturar a maior parte da variação.'''

import matplotlib.pyplot as plt
import numpy as np

# Calculando a variância acumulada
# array onde cada valor representa a variância total explicada até aquele componente principal.
var_acumulada = np.cumsum(pct)

# Configuração do gráfico
plt.figure(figsize=(14, 7))

# Gráfico de Barras para a Variância Explicada
plt.subplot(1, 2, 1)
plt.bar(range(1, len(pct) + 1), pct, alpha=0.7, align='center', color='blue')
plt.xlabel('Componentes Principais')
plt.ylabel('Variância Explicada (%)')
plt.title('Contribuição de Cada Componente Principal')
plt.xticks(range(1, len(pct) + 1))

# Gráfico de Linhas para a Variância Acumulada
plt.subplot(1, 2, 2)
plt.plot(range(1, len(var_acumulada) + 1), var_acumulada, marker='o', linestyle='--', color='red')
plt.xlabel('Componentes Principais')
plt.ylabel('Variância Acumulada (%)')
plt.title('Variância Acumulada por Componentes Principais')
plt.xticks(range(1, len(var_acumulada) + 1))
plt.ylim(0, 1.05)

# Exibindo os gráficos
plt.tight_layout()
plt.show()
```

```python
# Transformando os dados para obter os scores dos componentes principais

#  matriz onde cada linha corresponde aos dados originais representados em termos dos componentes principais.
pca_scores = pca.transform(retornos)

# Convertendo os scores para um DataFrame para facilitar a manipulação
pca_scores_df = pd.DataFrame(pca_scores, columns=[f'C{i+1}' for i in range(n)])

# Plotando os scores dos componentes principais ao longo do tempo
plt.figure(figsize=(14, 7))
for i in range(n):
    plt.plot(pca_scores_df.index, pca_scores_df[f'C{i+1}'], label=f'Componente {i+1}')
plt.xlabel('Data')
plt.ylabel('Score do Componente Principal')
plt.title('Scores dos Componentes Principais ao Longo do Tempo')
plt.legend()
```

```python
'''
DataFrame com os retornos projetados nos componentes principais.
components_returns, é uma matriz contendo os dados projetados no espaço dos componentes principais.
'''

X = np.asarray(retornos)
components_returns = X.dot(pca_components.T) # no PCA os componentes são definidos por linha, mas para multiplicação correta, eles precisam ser organizados como colunas.

components_returns = pd.DataFrame(
    columns=[f'C{i+1}' for i in range(n)] ,
    index=retornos.index,
    data=components_returns
)

components_returns
```

```python
'''
as exposições (ou pesos) dos ativos originais em cada um dos componentes principais obtidos pela PCA.
components_exposures mostra a exposição de cada ativo a cada um dos componentes principais. 
Isso significa que:
-As linhas do DataFrame representam os ativos originais.
-As colunas do DataFrame correspondem aos componentes principais (C1, C2, ..., Cn).
-Os valores dentro do DataFrame indicam a magnitude com que cada ativo contribui para o respectivo componente principal.
'''

components_exposures = pd.DataFrame(
    index = [f'C{i+1}' for i in range(n)],
    columns = retornos.columns, 
    data = pca_components #Cada linha de pca_components representa um componente principal, e seus valores indicam as contribuições (ou exposições) de cada ativo nesse componente.
    ).T

components_exposures 
```
